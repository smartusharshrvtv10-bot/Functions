 1. What is the difference between a function and a method in Python?
  - A function is a block of reusable code that is defined independently. A method is a function that is associated with an object and is called on that object (it's defined inside a class).
Example:
Function: len([1, 2, 3])
Method: [1, 2, 3].append(4)

 2. Explain the concept of function arguments and parameters in Python.
  - Parameters are the names defined in the function definition that specify what inputs the function can accept. Arguments are the actual values passed to the function when it is called.
  Example:
'a' and 'b' are parameters
def add(a, b):
    return a + b
5 and 3 are arguments
result = add(5, 3)

3. What are the different ways to define and call a function in Python?
 - A function is defined using the def keyword, followed by the function name, parameters in parentheses, and a colon, with the body indented. A function is called by using its name followed by the arguments in parentheses.
 Example:
 Defining
def greet(name):
    print(f"Hello, {name}")

Calling

4. What is the purpose of the `return` statement in a Python function?
 - The return statement is used to exit a function and pass a value (or object) back to the caller. If no value is specified, or if return is omitted, the function returns None.
 Example:
 def square(x):
    return x * x # Returns the calculated value

 5. What are iterators in Python and how do they differ from iterables?
  - An iterable is an object capable of returning its members one at a time (e.g., lists, tuples, strings). An iterator is an object that represents a stream of data, it is responsible for remembering where it is during iteration and has the special methods _iter() and __next_() (or next()). Iterators are obtained from iterables.
How they differ: You can iterate over an iterable multiple times. An iterator is "consumed" as you iterate and can typically only be used once.
Example:
Iterable: my_list = [1, 2, 3]
Iterator: my_iter = iter(my_list)

6. Explain the concept of generators in Python and how they are defined.
 - A generator is a simple way to create iterators. They are defined like normal functions but use the yield keyword instead of return to produce a sequence of results. When yield is encountered, the generator's state is saved, and execution can resume from that point later.
How they are defined: By using def and including one or more yield statements.
Example:
def count_up_to(n):
    i = 1
    while i <= n:
        yield i # Makes it a generator
        i += 1

7. What are the advantages of using generators over regular functions?
 - The main advantages of generators are:
Memory Efficiency: They produce items one at a time (lazy evaluation) only when requested, saving memory, especially for very large sequences.
Performance: They start generating values immediately, reducing initial load time.
Simplicity: They simplify the process of creating iterators.

8. What is a lambda function in Python and when is it typically used?
 - A lambda function is a small, anonymous (nameless) inline function defined using the lambda keyword. It can take any number of arguments but can only have one expression.
Syntax: lambda arguments: expression
Typical Use: They are typically used for short-term operations where a full function definition (def) is unnecessary, often as an argument to higher-order functions like map(), filter(), and sorted().
Example:
squared = lambda x: x * x

9. Explain the purpose and usage of the `map()` function in Python.
 - The map() function is used to apply a given function to all items in an input iterable (like a list) and return a map object (which is an iterator) that yields the results.
Usage: map(function, iterable)

 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - map(): Transforms each element in an array and returns a new array of the same length, containing the transformed elements. It applies a function to every element individually.
  filter(): Creates a new array containing only the elements from the original array that satisfy a specified condition. Elements for which the provided function returns true are included.
  reduce(): Executes a reducer function on each element of the array, resulting in a single output value. It iterates through the array and accumulates a single result based on a provided function.

```python
#Ques1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even_numbers(numbers):
    total = 0
    for number in numbers:
        if number % 2 == 0:
            total += number
    return total
```

```python
#Ques4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):
    """Checks if a given number is prime or not."""
    if number <= 1:
        return False
    '''Check for factors from 2 up to the square root of the number'''
    i = 2
    while i * i <= number:
        if number % i == 0:
            return False
        i += 1
    return True
```

```python
#Ques6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

```

```python
#Ques7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
```

```python
#Ques8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
def sort_tuples_by_second_element(list_of_tuples):
    list_of_tuples.sort(key=lambda x: x[1])
    return list_of_tuples
```

```python
#Ques9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def convert_temperatures(celsius_list):
    fahrenheit_iterator = map(celsius_to_fahrenheit, celsius_list)
    return list(fahrenheit_iterator) # Convert the map object to a list for printing
```

```python
#Ques10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'

    consonants_iterator = filter(lambda c: c not in vowels, input_string)

    return "".join(consonants_iterator)
```
